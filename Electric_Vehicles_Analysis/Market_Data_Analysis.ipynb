```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing the dataset
data=pd.read_csv('ev_data_US.csv')
data.head()
```

```python
# Adoption of Electric Vehicles over time
adoption_by_year = data['Model Year'].value_counts().sort_index()
```

```python
# Top 10 Counties with the highest number of Electric Vehicles
top_10_counties = data['County'].value_counts().head(10)
```

```python
# Top 10 Cities with the highest number of Electric Vehicles
top_10_cities = data['City'].value_counts().head(10)
```

```python
# Electric Vehicle Type
ev_type = data['Electric Vehicle Type'].value_counts()
```

```python
# Top 10 Electric Vehicle Manufacturer
top_10_ev_manufacturer = data['Make'].value_counts().head(10)
```

```python
# Top 10 Electric Vehicle Model
top_10_ev_model = data['Model'].value_counts().head(10)
```

```python
# Top 10 Electric Vehicle by Average Range
top_10_ev_by_avg_range = data.groupby('Model')['Electric Range'].mean().sort_values(ascending=False).head(10)
```

```python
# Clean Alternative Fuel Vehicle (CAFV) Eligibility
cafv_eligibility = data['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].value_counts()
```

```python
# Plotting the Adoption of Electric Vehicles over time
plt.figure(figsize=(16,8))
sns.set_style('darkgrid')
sns.barplot(x=adoption_by_year.index,
            y=adoption_by_year.values,
            palette='viridis',
            hue=adoption_by_year.index,
            legend=False)
plt.title('Adoption of Electric Vehicles over time')
plt.xlabel('Year')
plt.ylabel('Number of Electric Vehicles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plotting the Top 10 Counties with the highest number of Electric Vehicles
plt.figure(figsize=(16,8))
sns.set_style('darkgrid')
sns.barplot(x=top_10_counties.values,
            y=top_10_counties.index,
            palette='viridis',
            hue=top_10_counties.index,
            legend=False)
plt.title('Top 10 Counties with the highest number of Electric Vehicles')
plt.xlabel('County')
plt.ylabel('Number of Electric Vehicles')
plt.tight_layout()
plt.show()
```

```python
# Plotting the Top 10 Cities with the highest number of Electric Vehicles
plt.figure(figsize=(16,8))
sns.set_style('darkgrid')
sns.barplot(x=top_10_cities.values,
            y=top_10_cities.index,
            palette='viridis',
            hue=top_10_cities.index,
            legend=False)
plt.title('Top 10 Cities with the highest number of Electric Vehicles')
plt.xlabel('City')
plt.ylabel('Number of Electric Vehicles')
plt.tight_layout()
plt.show()
```

```python
# Plotting the Electric Vehicle Type by Count
plt.figure(figsize=(16,8))
sns.set_style('darkgrid')
sns.barplot(x=ev_type.values,
            y=ev_type.index,
            palette='viridis',
            hue=ev_type.index,
            legend=False)
plt.title('Electric Vehicle Type by Count')
plt.xlabel('Electric Vehicle Type')
plt.ylabel('Number of Electric Vehicles')
plt.tight_layout()
plt.show()
```

```python
# Plotting the Top 10 Electric Vehicle Manufacturer
plt.figure(figsize=(16,8))
sns.set_style('darkgrid')
sns.barplot(x=top_10_ev_manufacturer.values,
            y=top_10_ev_manufacturer.index,
            palette='viridis',
            hue=top_10_ev_manufacturer.index,
            legend=False)
plt.title('Top 10 Electric Vehicle Manufacturer by Count')
plt.xlabel('Manufacturer')
plt.ylabel('Number of Electric Vehicles')
plt.tight_layout()
plt.show()
```

```python
# Plotting the Top 10 Electric Vehicle Model
plt.figure(figsize=(16,8))
sns.set_style('darkgrid')
sns.barplot(x=top_10_ev_model.values,
            y=top_10_ev_model.index,
            palette='viridis',
            hue=top_10_ev_model.index,
            legend=False)
plt.title('Top 10 Electric Vehicle Model by Count')
plt.xlabel('Model')
plt.ylabel('Number of Electric Vehicles')
plt.tight_layout()
plt.show()
```

```python
# Plotting the Top 10 Electric Vehicle by Average Range
plt.figure(figsize=(16,8))
sns.set_style('darkgrid')
sns.barplot(x=top_10_ev_by_avg_range.values,
            y=top_10_ev_by_avg_range.index,
            palette='viridis',
            hue=top_10_ev_by_avg_range.index,
            legend=False)
plt.title('Top 10 Electric Vehicle by Average Range')
plt.xlabel('Model')
plt.ylabel('Average Electric Range')
plt.tight_layout()
plt.show()
```

```python
# Plotting the Clean Alternative Fuel Vehicle (CAFV) Eligibility
plt.figure(figsize=(16,8))
sns.set_style('darkgrid')
sns.barplot(x=cafv_eligibility.index,
            y=cafv_eligibility.values,
            palette='viridis',
            hue=cafv_eligibility.index,
            legend=False)
plt.title('Clean Alternative Fuel Vehicle (CAFV) Eligibility')
plt.xlabel('CAFV Eligibility')
plt.ylabel('Number of Electric Vehicles')
plt.tight_layout()
plt.show()
```
